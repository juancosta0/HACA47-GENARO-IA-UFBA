```python
import requests
import pandas as pd

# Endpoint público do ngrok (substitua pela URL gerada pelo ngrok)
NGROK_URL = "https://9498-34-31-15-45.ngrok-free.app"

# Nome do modelo no Ollama
MODEL_NAME = "llama3.2"

# Caminho do arquivo Excel com as perguntas
EXCEL_FILE = "perguntas.xlsx"

# Função para enviar perguntas ao Ollama
def enviar_pergunta(pergunta):
    try:
        # URL para o endpoint do Ollama via ngrok
        url = f"{NGROK_URL}/api/v1/completions"
        payload = {
            "model": MODEL_NAME,
            "prompt": pergunta
        }
        headers = {
            "Content-Type": "application/json"
        }

        # Enviar a requisição POST
        response = requests.post(url, json=payload, headers=headers)

        # Verificar se a resposta foi bem-sucedida
        if response.status_code == 200:
            data = response.json()
            return data.get("completion", "Sem resposta")
        else:
            return f"Erro na requisição: {response.status_code} - {response.text}"
    except Exception as e:
        return f"Erro ao conectar ou processar: {e}"

# Carregar o arquivo Excel
try:
    df = pd.read_excel(EXCEL_FILE)
except Exception as e:
    print(f"Erro ao carregar o arquivo Excel: {e}")
    exit()

# Coluna de perguntas e respostas
if "Pergunta" not in df.columns:
    print("O arquivo Excel deve conter uma coluna chamada 'Pergunta'.")
    exit()

# Adicionar uma coluna para armazenar as respostas da IA
df["Resposta IA"] = ""

# Processar cada pergunta e enviar para a IA
for idx, row in df.iterrows():
    pergunta = row["Pergunta"]
    print(f"Enviando pergunta: {pergunta}")
    resposta = enviar_pergunta(pergunta)
    print(f"Resposta: {resposta}")
    df.at[idx, "Resposta IA"] = resposta

# Salvar os resultados em um novo arquivo Excel
output_file = "resultados_avaliacao.xlsx"
df.to_excel(output_file, index=False)
print(f"Resultados salvos em: {output_file}")
```
