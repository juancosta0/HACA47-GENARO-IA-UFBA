#Script do rag (Perguntas abertas)

```python
import json
import os
import fitz  # PyMuPDF
import requests
from google.colab import files

# 🔹 Configuração do servidor Ollama via ngrok
OLLAMA_URL = "http://localhost:11434/api/generate"
MODEL_NAME = "llama3.2"

# 🔹 Função para fazer upload no Google Colab
def upload_files(prompt_message):
    print(prompt_message)
    uploaded = files.upload()
    return uploaded

# 🔹 Função para extrair texto dos PDFs
def extract_text_from_pdfs(pdf_folder):
    text_data = []

    for file_name in os.listdir(pdf_folder):
        if file_name.endswith(".pdf"):
            file_path = os.path.join(pdf_folder, file_name)

            with fitz.open(file_path) as pdf:
                text = ""
                for page in pdf:
                    text += page.get_text("text") + "\n"
                text_data.append(text)

    return " ".join(text_data)  # Junta todo o texto dos PDFs

# 🔹 Função para perguntar ao modelo via API do Ollama
def ask_model(question, context, model=MODEL_NAME):
    payload = {
        "model": model,
        "prompt": f"Com base neste contexto, responda de forma direta:\n\n{context}\n\nPergunta: {question}",
        "stream": False
    }
    response = requests.post(OLLAMA_URL, json=payload)

    if response.status_code == 200:
        return response.json().get("response", "Erro na resposta")
    else:
        return f"Erro: {response.status_code} - {response.text}"

# 🔹 Função para avaliar o modelo
def evaluate_model(json_path, pdf_folder, model=MODEL_NAME):
    # Carregar perguntas
    with open(json_path, 'r', encoding='utf-8') as file:
        questions = json.load(file)

    # Extrair texto dos PDFs como contexto
    pdf_text = extract_text_from_pdfs(pdf_folder)

    correct_count = 0
    total_questions = sum(len(v) for v in questions.values())
    respostas_ia = []

    for page, question_list in questions.items():
        for question_data in question_list:
            question = question_data["Pergunta"]
            correct_answer = str(question_data["Resposta Correta"]).strip().lower()

            # Perguntar ao modelo com contexto
            model_answer = ask_model(question, pdf_text, model=model).strip().lower()

            print(f"\n🔹 Pergunta: {question}")
            print(f"✅ Resposta correta: {correct_answer}")
            print(f"🤖 Resposta da IA: {model_answer}")

            # Salvar resposta para análise posterior
            respostas_ia.append({
                "Pergunta": question,
                "Resposta_Correta": correct_answer,
                "Resposta_IA": model_answer
            })

            # Comparação exata (pode ser melhorada)
            if model_answer == correct_answer:
                correct_count += 1

    # Salvar respostas geradas
    with open("respostas_IA.json", "w", encoding="utf-8") as output_file:
        json.dump(respostas_ia, output_file, indent=4, ensure_ascii=False)

    # Exibir taxa de acerto
    accuracy = (correct_count / total_questions) * 100
    print(f"\n🎯 Porcentagem de acertos: {accuracy:.2f}%")
    return accuracy

# 🔹 Execução no Google Colab
if __name__ == "__main__":
    # Upload dos arquivos
    upload_files("Faça o upload do arquivo JSON com as perguntas...")
    json_filename = "perguntas.json"

    upload_files("Agora faça o upload dos arquivos PDF para serem usados no RAG...")
    PDF_FOLDER = "pdfs"
    os.makedirs(PDF_FOLDER, exist_ok=True)

    # Mover PDFs para a pasta correta
    for filename in os.listdir():
        if filename.endswith(".pdf"):
            os.rename(filename, os.path.join(PDF_FOLDER, filename))

    print(f"📂 Todos os PDFs foram salvos em '{PDF_FOLDER}'.")

    # Executar a avaliação
    evaluate_model(json_filename, PDF_FOLDER, model=MODEL_NAME)
```

#Script do rag (Perguntas fechada)

```python
import json
import os
import fitz  # PyMuPDF
import requests
from google.colab import files

# 🔹 Configuração do servidor Ollama via ngrok
OLLAMA_URL = "http://localhost:11434/api/generate"
MODEL_NAME = "llama3.2"

# 🔹 Função para fazer upload no Google Colab
def upload_files(prompt_message):
    print(prompt_message)
    uploaded = files.upload()
    return uploaded

# 🔹 Função para extrair texto dos PDFs
def extract_text_from_pdfs(pdf_folder):
    text_data = []

    for file_name in os.listdir(pdf_folder):
        if file_name.endswith(".pdf"):
            file_path = os.path.join(pdf_folder, file_name)

            with fitz.open(file_path) as pdf:
                text = ""
                for page in pdf:
                    text += page.get_text("text") + "\n"
                text_data.append(text)

    return " ".join(text_data)  # Junta todo o texto dos PDFs

# 🔹 Função para perguntar ao modelo via API do Ollama
def ask_model(question, options, context, model=MODEL_NAME):
    options_text = "\n".join([f"{key}) {value}" for key, value in options.items()])
    prompt = f"Com base neste contexto, escolha a alternativa correta.\n\n{context}\n\nPergunta: {question}\n\nAlternativas:\n{options_text}\n\nResponda apenas com a letra correspondente, sem justificar."

    payload = {
        "model": model,
        "prompt": prompt,
        "stream": False
    }
    response = requests.post(OLLAMA_URL, json=payload)

    if response.status_code == 200:
        return response.json().get("response", "Erro na resposta").strip().lower()[0]  # Pega apenas a primeira letra
    else:
        return f"Erro: {response.status_code} - {response.text}"

# 🔹 Função para avaliar o modelo
def evaluate_model(json_path, pdf_folder, model=MODEL_NAME):
    # Carregar perguntas
    with open(json_path, 'r', encoding='utf-8') as file:
        questions = json.load(file)

    # Extrair texto dos PDFs como contexto
    pdf_text = extract_text_from_pdfs(pdf_folder)

    correct_count = 0
    total_questions = sum(len(v) for v in questions.values())
    respostas_ia = []

    for page, question_list in questions.items():
        for question_data in question_list:
            question = question_data["Pergunta"]
            options = question_data["Alternativas"]
            correct_answer = str(question_data["Resposta Correta"]).strip().lower()

            # Perguntar ao modelo com contexto
            model_answer = ask_model(question, options, pdf_text, model=model)

            print(f"\n🔹 Pergunta: {question}")
            for key, value in options.items():
                print(f"  {key}) {value}")
            print(f"✅ Resposta correta: {correct_answer}")
            print(f"🤖 Resposta da IA: {model_answer}")

            # Salvar resposta para análise posterior
            respostas_ia.append({
                "Pergunta": question,
                "Alternativas": options,
                "Resposta_Correta": correct_answer,
                "Resposta_IA": model_answer
            })

            # Comparação exata (pode ser melhorada)
            if model_answer == correct_answer:
                correct_count += 1

    # Salvar respostas geradas
    with open("respostas_IA.json", "w", encoding="utf-8") as output_file:
        json.dump(respostas_ia, output_file, indent=4, ensure_ascii=False)

    # Exibir taxa de acerto
    accuracy = (correct_count / total_questions) * 100
    print(f"\n🎯 Porcentagem de acertos: {accuracy:.2f}%")
    return accuracy

# 🔹 Execução no Google Colab
if __name__ == "__main__":
    # Upload dos arquivos
    upload_files("Faça o upload do arquivo JSON com as perguntas...")
    json_filename = "perguntas_fechadas.json"

    upload_files("Agora faça o upload dos arquivos PDF para serem usados no RAG...")
    PDF_FOLDER = "pdfs"
    os.makedirs(PDF_FOLDER, exist_ok=True)

    # Mover PDFs para a pasta correta
    for filename in os.listdir():
        if filename.endswith(".pdf"):
            os.rename(filename, os.path.join(PDF_FOLDER, filename))

    print(f"📂 Todos os PDFs foram salvos em '{PDF_FOLDER}'.")

    # Executar a avaliação
    evaluate_model(json_filename, PDF_FOLDER, model=MODEL_NAME)
```
